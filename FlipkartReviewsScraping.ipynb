```python
def get_reviews(search):
    reviews = []
    search_string = search.replace(" ", "")
    flipkart_url = "https://www.flipkart.com/search?q=" + search_string

    uClient = requests.get(flipkart_url)
    flipkart_html = bs(uClient.text, "html.parser")
    bigboxes = flipkart_html.findAll("div", {"class": "_1AtVbE col-12-12"})
    box = bigboxes[2]
    productLink = "https://www.flipkart.com" + box.div.div.div.a['href']
    prodRes = requests.get(productLink)
    prodRes.encoding = 'utf-8'
    prod_html = bs(prodRes.text, "html.parser")
    allReviews = prod_html.find_all('div', {'class': "col JOpGWq"})
    length = len(allReviews[0].findAll("a"))
    new_url = "https://www.flipkart.com" + allReviews[0].findAll("a")[length-1]['href']

    uClient = requests.get(new_url)
    helpfulReviewPage = uClient.text
    reviewPage_html = bs(helpfulReviewPage, "html.parser")
    helpfulbigboxes = reviewPage_html.findAll("div", {"class": "_27M-vq"})

    for i in range(len(helpfulbigboxes)-1):
        rev = helpfulbigboxes[i].div.div.findAll("div", {"class": "t-ZTKy"})[0].div.div.text
        rating = helpfulbigboxes[i].div.div.div.div.text
        mydict = {"Product": search_string, "Review": rev, "Sentiment": rating}
        reviews.append(mydict)

    x = new_url
    x = x.replace('&marketplace=FLIPKART', '&aid=overall&certifiedBuyer=false&sortOrder=NEGATIVE_FIRST')
    uClient = requests.get(x)
    negativeReviewPage = uClient.text
    reviewPage_html = bs(negativeReviewPage, "html.parser")
    negativebigboxes = reviewPage_html.findAll("div", {"class": "_27M-vq"})

    for i in range(len(negativebigboxes)-1):
        rev = negativebigboxes[i].div.div.findAll("div", {"class": "t-ZTKy"})[0].div.div.text
        rating = negativebigboxes[i].div.div.div.div.text
        mydict = {"Product": search_string, "Review": rev, "Sentiment": rating}
        reviews.append(mydict)

    return reviews
```
